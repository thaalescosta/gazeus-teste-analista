# **Distribuições de Probabilidade**

## **Distribuições Discretas**

### **1. Distribuição Binomial**
- **Quando usar**: Quando temos um número fixo de tentativas independentes, cada uma com dois resultados possíveis (sucesso ou fracasso), e a probabilidade de sucesso é constante.

- **Exemplo**: Número de caras em 10 lançamentos de uma moeda.

- **Parâmetros**: $n$ (número de tentativas), $p$ (probabilidade de sucesso).

- **Função de Probabilidade**:  
$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$


### **2. Distribuição de Poisson**
- **Quando usar**: Para modelar o número de eventos que ocorrem em um intervalo fixo de tempo ou espaço, quando esses eventos são raros e independentes.

- **Exemplo**: Número de chamadas recebidas por um call center em uma hora.

- **Parâmetros**: #\lambda$ (taxa média de ocorrência).

- **Função de Probabilidade**:  
$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

### **3. Distribuição Geométrica**
- **Quando usar**: Para modelar o número de tentativas até o primeiro sucesso em uma sequência de tentativas independentes.

- **Exemplo**: Número de lançamentos até obter a primeira cara em uma moeda.

- **Parâmetros**: $n$ (probabilidade de sucesso).

- **Função de Probabilidade**:  
  $$
  P(X = k) = (1-p)^{k-1} p
  $$

### **4. Distribuição Hipergeométrica**
- **Quando usar**: Para modelar a probabilidade de $k$ sucessos em uma amostra de tamanho $n$ retirada de uma população finita, sem reposição.

- **Exemplo**: Número de acertos ao escolher 5 cartas de um baralho sem reposição.

- **Parâmetros**: $N$ (tamanho da população), $K$ (número de sucessos na população), $n$ (tamanho da amostra).

- **Função de Probabilidade**:  
  $$
  P(X = k) = \frac{\binom{K}{k} \binom{N-K}{n-k}}{\binom{N}{n}}
  $$

---

## **Distribuições Contínuas**

### **1. Distribuição Normal**
- **Quando usar**: Para modelar variáveis que se distribuem simetricamente em torno de uma média, com uma forma de "campanha". A distribuição normal é amplamente usada em estatísticas.

- **Exemplo**: Altura das pessoas em uma população.

- **Parâmetros**: $\mu$ (média), $\sigma$ (desvio padrão).

- **Função de Probabilidade**:  
  $$
  f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
  $$

### **2. Distribuição Exponencial**
- **Quando usar**: Para modelar o tempo entre eventos em um processo de Poisson, ou seja, a distribuição do tempo até o primeiro evento.

- **Exemplo**: Tempo até o próximo cliente chegar a um caixa.

- **Parâmetros**: $\lambda$ (taxa de ocorrência).

- **Função de Probabilidade**:  
  $$
  f(x) = \lambda e^{-\lambda x}, \quad x \geq 0
  $$

### **3. Distribuição Uniforme (Contínua)**
- **Quando usar**: Quando todos os valores de uma variável dentro de um intervalo são igualmente prováveis.

- **Exemplo**: Lançamento de um dado perfeito.

- **Parâmetros**: $a$ (início do intervalo), $b$ (fim do intervalo).

- **Função de Probabilidade**:  
  $$
  f(x) = \frac{1}{b-a}, \quad a \leq x \leq b
  $$

### **4. Distribuição Beta**
- **Quando usar**: Para modelar variáveis contínuas que são limitadas entre 0 e 1, como proporções ou probabilidades.

- **Exemplo**: Probabilidade de um site ser acessado, dada uma proporção de acessos bem-sucedidos.

- **Parâmetros**: $\alpha$ e $\beta$ (formas).

- **Função de Probabilidade**:  
  $$
  f(x; \alpha, \beta) = \frac{x^{\alpha - 1} (1 - x)^{\beta - 1}}{B(\alpha, \beta)}
  $$

### **5. Distribuição Gamma**
- **Quando usar**: Para modelar o tempo até $k$ eventos em um processo de Poisson.

- **Exemplo**: Tempo até 5 chamadas em um call center.

- **Parâmetros**: $k$ (forma), $\lambda$ (taxa).
- **Função de Probabilidade**:  
  $$
  f(x; k, \lambda) = \frac{\lambda^k x^{k-1} e^{-\lambda x}}{(k-1)!}, \quad x \geq 0
  $$

---

## **Distribuições para Grandes Amostras / Aproximações**

### **1. Distribuição Normal (para grandes amostras)**
- **Quando usar**: Aproximação para distribuições binomial e de Poisson, quando o tamanho da amostra é grande.

- **Exemplo**: Aproximação de uma distribuição binomial com $n$ grande.

- **Parâmetros**: $\mu = np$, $\sigma = \sqrt{np(1-p)}$ para a binomial.

### **2. Distribuição t de Student**
- **Quando usar**: Para testar hipóteses sobre a média de uma população quando o tamanho da amostra é pequeno e a variância é desconhecida.

- **Exemplo**: Teste $t$ para comparar médias de dois grupos com pequenas amostras.

- **Parâmetros**: Graus de liberdade $df$.

### **3. Distribuição Qui-quadrado**
- **Quando usar**: Para testar a independência ou ajuste de uma tabela de contingência, ou para testar a variância de uma população.

- **Exemplo**: Teste de qui-quadrado para ajustar a distribuição observada a uma distribuição esperada.

- **Parâmetros**: Graus de liberdade $df$.

### **4. Distribuição F**
- **Quando usar**: Para comparar variâncias entre dois grupos ou realizar uma análise de variância (ANOVA).

- **Exemplo**: Teste F para comparar variâncias de dois grupos em um estudo.

- **Parâmetros**: Graus de liberdade $df_1$ e $df_2$.

---

## **Resumo de Como Identificar a Distribuição**
- **Binomial**: Experimentos com $n$ tentativas e dois resultados possíveis.

- **Poisson**: Número de eventos em um intervalo de tempo ou espaço.

- **Normal**: Variáveis contínuas simetricamente distribuídas, frequentemente usadas para modelar dados reais.

- **Exponencial**: Tempo até o primeiro evento em um processo de Poisson.

- **Gamma/Beta**: Modelagem de tempo até $k$ eventos ou proporções.

# Guia para Identificar a Melhor Distribuição de Probabilidade para um Dataset

## Introdução

Este guia auxilia na identificação da distribuição de probabilidade que melhor representa um conjunto de dados, seja ele contínuo ou discreto.

## Distribuições Discretas

### **1. Distribuição Binomial**
- **Quando usar**: Quando temos um número fixo de tentativas independentes, cada uma com dois resultados possíveis (sucesso ou fracasso), e a probabilidade de sucesso é constante.

- **Exemplo**: Número de caras em 10 lançamentos de uma moeda.

- **Parâmetros**: $n$ (número de tentativas), $p$ (probabilidade de sucesso).

- **Função de Probabilidade**:  
$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$

### **2. Distribuição de Poisson**
- **Quando usar**: Para modelar o número de eventos que ocorrem em um intervalo fixo de tempo ou espaço, quando esses eventos são raros e independentes.

- **Exemplo**: Número de chamadas recebidas por uma central de atendimento em uma hora.

- **Parâmetros**: $\lambda$ (taxa média de ocorrência).

- **Função de Probabilidade**:  
$$P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}$$

### **3. Distribuição Geométrica**
- **Quando usar**: Para modelar o número de tentativas necessárias até o primeiro sucesso em uma série de tentativas independentes.

- **Exemplo**: Número de lançamentos de um dado até obter um 6.

- **Parâmetros**: $p$ (probabilidade de sucesso em cada tentativa).

- **Função de Probabilidade**:  
$$P(X = k) = p(1-p)^{k-1}$$

### **4. Distribuição Binomial Negativa**
- **Quando usar**: Para modelar o número de fracassos antes de um número fixo de sucessos em uma série de tentativas independentes.

- **Exemplo**: Número de falhas antes de obter 5 sucessos em um processo de controle de qualidade.

- **Parâmetros**: $r$ (número de sucessos desejados), $p$ (probabilidade de sucesso em cada tentativa).

- **Função de Probabilidade**:  
$$P(X = k) = \binom{k+r-1}{k} p^r (1-p)^k$$

### **5. Distribuição Hipergeométrica**
- **Quando usar**: Para modelar o número de sucessos em uma amostra de tamanho fixo, retirada sem reposição de uma população finita.

- **Exemplo**: Número de itens defeituosos em uma amostra de 10 itens retirados de um lote de 100.

- **Parâmetros**: $N$ (tamanho da população), $K$ (número de sucessos na população), $n$ (tamanho da amostra).

- **Função de Probabilidade**:  
$$P(X = k) = \frac{\binom{K}{k} \binom{N-K}{n-k}}{\binom{N}{n}}$$

## Distribuições Contínuas

### **1. Distribuição Normal (Gaussiana)**
- **Quando usar**: Para modelar fenômenos naturais e sociais que tendem a se agrupar simetricamente em torno de um valor médio.

- **Exemplo**: Altura de adultos em uma população.

- **Parâmetros**: $\mu$ (média), $\sigma$ (desvio padrão).

- **Função de Densidade de Probabilidade**:  
$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

### **2. Distribuição Log-Normal**
- **Quando usar**: Para modelar variáveis positivas que são produto de muitos fatores independentes.

- **Exemplo**: Renda pessoal, preços de ações.

- **Parâmetros**: $\mu$ (média do logaritmo dos dados), $\sigma$ (desvio padrão do logaritmo dos dados).

- **Função de Densidade de Probabilidade**:  
$$f(x) = \frac{1}{x\sigma \sqrt{2\pi}} e^{-\frac{(\ln x-\mu)^2}{2\sigma^2}}$$

### **3. Distribuição Exponencial**
- **Quando usar**: Para modelar o tempo entre eventos em um processo de Poisson ou a vida útil de componentes.

- **Exemplo**: Tempo entre chegadas de clientes em uma fila.

- **Parâmetros**: $\lambda$ (taxa).

- **Função de Densidade de Probabilidade**:  
$$f(x) = \lambda e^{-\lambda x}$$

### **4. Distribuição Weibull**
- **Quando usar**: Para modelar tempo até falha em análises de confiabilidade, especialmente quando a taxa de falha varia com o tempo.

- **Exemplo**: Vida útil de componentes eletrônicos.

- **Parâmetros**: $k$ (parâmetro de forma), $\lambda$ (parâmetro de escala).

- **Função de Densidade de Probabilidade**:  
$$f(x) = \frac{k}{\lambda} (\frac{x}{\lambda})^{k-1} e^{-(x/\lambda)^k}$$

### **5. Distribuição Gamma**
- **Quando usar**: Para modelar o tempo até a ocorrência de um número específico de eventos em um processo de Poisson.

- **Exemplo**: Quantidade total de chuva acumulada em um período.

- **Parâmetros**: $k$ (parâmetro de forma), $\theta$ (parâmetro de escala).

- **Função de Densidade de Probabilidade**:  
$$f(x) = \frac{x^{k-1} e^{-x/\theta}}{\theta^k \Gamma(k)}$$

### **6. Distribuição Beta**
- **Quando usar**: Para modelar proporções e probabilidades, especialmente em contextos bayesianos.

- **Exemplo**: Proporção de tempo que um sistema fica operacional.

- **Parâmetros**: $\alpha$ (primeiro parâmetro de forma), $\beta$ (segundo parâmetro de forma).

- **Função de Densidade de Probabilidade**:  
$$f(x) = \frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha,\beta)}$$

# Guide to Choosing the Right Probability Distribution

## 1. Is Your Data Discrete or Continuous?
First, determine whether your data is:
- **Discrete** (countable, integer-valued, e.g., number of storms, occurrences)
- **Continuous** (measurable, real-valued, e.g., temperature, rainfall amounts)

---

## 2. Choosing a Distribution for Discrete Data

### Key Discrete Distributions

| Distribution         | When to Use | Key Characteristics | Example Applications |
|----------------------|-------------|----------------------|----------------------|
| **Bernoulli**        | When modeling single yes/no outcomes with constant probability | Single trial, two possible outcomes | Rain/no rain on a given day |
| **Binomial**         | Counts of successes in fixed number of independent trials with constant success probability | Variance = n*p*(1-p) | Number of freezing events in 10 years |
| **Geometric**        | Number of trials needed to get first success in repeated independent trials | Memoryless property | Waiting time for first hurricane |
| **Negative Binomial**| Number of trials needed to get k successes | Generalization of geometric distribution | Dry days before 3 rainfall events |
| **Poisson**          | Counts of rare events occurring at constant rate in fixed interval | Mean = variance | Annual hurricane landfalls |

---

## 3. Choosing a Distribution for Continuous Data

### Key Continuous Distributions

| Distribution       | When to Use | Key Characteristics | Example Applications |
|--------------------|-------------|----------------------|----------------------|
| **Gaussian**       | When data is symmetric and approximately bell-shaped | Defined by μ (mean) and σ (standard deviation) | Monthly temperatures |
| **Lognormal**      | When data is positive and right-skewed, especially for multiplicative processes | log(X) follows normal distribution | Precipitation amounts |
| **Exponential**    | For time between independent events occurring at constant rate | Memoryless property, constant hazard rate | Time between storms |
| **Gamma**          | For positive, right-skewed data with flexible shape | Generalization of exponential distribution | Rainfall totals |
| **Weibull**        | For time-to-failure data with non-constant hazard rate | Flexible shape parameter | Wind speeds |
| **Beta**           | For proportions or percentages bounded between 0 and 1 | Flexible shape on bounded interval | Humidity percentages |
| **Extreme-Value**  | For modeling maximum/minimum values | Heavy-tailed distributions | Flood levels |
| **Mixture Models** | When data comes from multiple subpopulations | Weighted sum of component distributions | El Niño vs non-El Niño temps |

---

## 4. Practical Implementation Steps

```python
## Corrected Python Implementation for Distribution Fitting

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Simulate overdispersed count data
mu, size = 5, 1.2  # size = dispersion (smaller = more overdispersion)
data = stats.nbinom.rvs(n=size, p=size/(size+mu), size=1000)

# Correct fitting method:
def fit_poisson(data):
    return np.mean(data)  # MLE for Poisson is just the sample mean

def fit_negative_binomial(data):
    # Method of moments estimation
    mean = np.mean(data)
    var = np.var(data)
    p = mean/var
    n = (mean*p)/(1-p)
    return n, p

poisson_mu = fit_poisson(data)
nb_n, nb_p = fit_negative_binomial(data)

print(f"Poisson μ: {poisson_mu:.2f}")
print(f"Negative Binomial: n={nb_n:.2f}, p={nb_p:.2f}")
print(f"Implied dispersion: {(1-nb_p)/nb_p:.2f}")

# Visualization
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# Histogram with fits
ax[0].hist(data, bins=30, density=True, alpha=0.6)
x = np.arange(0, 20)
ax[0].plot(x, stats.poisson.pmf(x, poisson_mu), 'r-', label='Poisson')
ax[0].plot(x, stats.nbinom.pmf(x, nb_n, nb_p), 'g-', label='Negative Binomial')
ax[0].legend()

# QQ Plot
stats.probplot(data, dist=stats.nbinom(nb_n, nb_p), plot=ax[1])
ax[1].set_title('QQ Plot vs Negative Binomial')
plt.tight_layout()
```

### Negative Binomial and Related Distributions

| Distribution         | When to Use | Relationship to Other Distributions | Key Parameters | Python Implementation |
|----------------------|-------------|--------------------------------------|----------------|-----------------------|
